# annotation.jsonl　の分析

```python
annotation_path = "/workspace/data/annotations.jsonl"

# jsonlファイルの読み込み
import json
with open(annotation_path, "r") as f:
    annotations = [json.loads(line) for line in f]
    
# json ファイルの内容を確認
print(json.dumps(annotations[0], indent=4))
```

```python
# json から selected_class と selected_subclass を抽出
selected_classes = [ann["selected_class"] for ann in annotations]
selected_subclasses = [ann["selected_subclass"] for ann in annotations]

# uniqueにする
unique_classes = set(selected_classes)
unique_subclasses = set(selected_subclasses)

print(f"num of unique classes: {len(unique_classes)}")
print(f"num of unique subclasses: {len(unique_subclasses)}")
print("-"*30)

for i, cls in enumerate(unique_classes):
    print(f"Class {i}: {cls}")

print("-"*30)    

for i, cls in enumerate(unique_subclasses):
    print(f"Subclass {i}: {cls}")    
```

```python
# unique class　に紐づく subclass を確認.
class_to_subclass = {}
for cls, subcls in zip(selected_classes, selected_subclasses):
    if cls not in class_to_subclass:
        class_to_subclass[cls] = set()
    class_to_subclass[cls].add(subcls)
    
for cls, subcls_set in class_to_subclass.items():
    print(f"Class: {cls}")
    print(f"Subclasses: {subcls_set}")
    print("-"*30)
```

```python
# selected_class と instruction の組み合わせを確認
class_instruction_pairs = set()
for ann in annotations:
    pair = (ann["selected_class"], ann["instruction"])
    class_instruction_pairs.add(pair)
    
# selected_subclass と instruction の組み合わせを確認
subclass_instruction_pairs = set()
for ann in annotations:
    pair = (ann["selected_subclass"], ann["instruction"])
    subclass_instruction_pairs.add(pair)
    
# paris の中身を表示
print("Class - Instruction pairs:")
for pair in class_instruction_pairs:
    print(pair)

print("\nSubclass - Instruction pairs:")
for pair in subclass_instruction_pairs:
    print(pair)
```
