# change_color: LLM-based Color Target Extraction + cv2 Color Change

このノートブックは以下を目的にしています。

- instruction テキストから LLM で「何の色を何色に変えるか」を抽出
- 抽出した対象オブジェクト（例：shirt, tie）と色情報を基に、GroundingDINO + SAM でマスク生成
- cv2 で HSV 色空間をシフトして色変更を実行
- 後で `.py` に切り出しやすいよう、関数を分割して構成

```python
# Execution example: gradual color change (e.g., hair to red)
out_dir = Path('/workspace/logs/submit/change_color')
out_dir.mkdir(parents=True, exist_ok=True)

# Use available video
input_video = '/workspace/data/videos/bkwLxcfcKTY_1_71to281.mp4'
output_video = str(out_dir / 'output__hair_to_red_gradual.mp4')
target_prompt = 'hair .'
target_color = 'red'
max_frames = 60

result = run_change_color_gradual_pipeline(
    input_video=input_video,
    output_video=output_video,
    target_prompt=target_prompt,
    target_color=target_color,
    max_frames=max_frames,
)

print('\n=== Gradual Color Change Result ===')
for key, value in result.items():
    print(f'{key}: {value}')

# Show before/after comparison
show_before_after(input_video, output_video)
```

```python
from __future__ import annotations

from dataclasses import dataclass, field
from pathlib import Path
import sys
import json
import re

import cv2
import numpy as np
from tqdm.auto import tqdm

# Project imports
sys.path.append('/workspace/src')
sys.path.append('/workspace/src/utils')

from postprocess.detectors import detect_all_boxes, get_sam_mask_from_box
from video_utility import load_video, write_video, show_before_after


@dataclass
class ChangeColorConfig:
    """Configuration for LLM-based color change pipeline."""
    input_video: str
    output_video: str
    instruction: str
    dino_box_threshold: float = 0.30
    dino_text_threshold: float = 0.25
    max_frames: int | None = None
    hue_shift: int = 30
    saturation_scale: float = 1.0
    value_scale: float = 1.0


def color_name_to_hue_range(color_name: str) -> tuple[int, int]:
    """Map color name to HSV hue range (0-180 for OpenCV)."""
    color_ranges = {
        'red': (0, 10), 'orange': (10, 25), 'yellow': (25, 35),
        'green': (35, 85), 'cyan': (85, 95), 'blue': (95, 130),
        'purple': (130, 145), 'pink': (145, 170), 'magenta': (145, 170),
    }
    normalized = color_name.lower().strip()
    return color_ranges.get(normalized, (95, 130))


def estimate_hue_shift(from_color: str, to_color: str) -> int:
    """Estimate hue shift value based on color names."""
    from_range = color_name_to_hue_range(from_color)
    to_range = color_name_to_hue_range(to_color)
    from_hue = (from_range[0] + from_range[1]) // 2
    to_hue = (to_range[0] + to_range[1]) // 2
    shift = to_hue - from_hue
    return (shift + 180) % 180


def apply_color_change(
    frame_bgr: np.ndarray,
    target_mask_u8: np.ndarray,
    hue_shift: int,
    saturation_scale: float = 1.0,
    value_scale: float = 1.0,
) -> np.ndarray:
    """Apply color change in HSV space on masked region only."""
    hsv = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2HSV).astype(np.float32)
    mask_float = target_mask_u8.astype(np.float32)
    hsv[..., 0] = (hsv[..., 0] + hue_shift) % 180.0
    hsv[..., 1] = np.clip(hsv[..., 1] * saturation_scale, 0.0, 255.0)
    hsv[..., 2] = np.clip(hsv[..., 2] * value_scale, 0.0, 255.0)
    hsv_out = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2HSV).astype(np.float32)
    for c in range(3):
        hsv_out[..., c] = hsv_out[..., c] * (1.0 - mask_float) + hsv[..., c] * mask_float
    return cv2.cvtColor(hsv_out.astype(np.uint8), cv2.COLOR_HSV2BGR)


def estimate_target_mask_gdino_sam(
    frame_bgr: np.ndarray,
    target_prompt: str,
    dino_box_threshold: float = 0.30,
    dino_text_threshold: float = 0.25,
) -> np.ndarray:
    """Estimate target object mask using GroundingDINO + SAM."""
    frame_rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    boxes = detect_all_boxes(
        frame_rgb,
        text_prompt=target_prompt,
        box_threshold=dino_box_threshold,
        text_threshold=dino_text_threshold,
    )
    if not boxes:
        return np.zeros(frame_bgr.shape[:2], dtype=np.uint8)
    best_box = max(boxes, key=lambda b: (b[2] - b[0]) * (b[3] - b[1]))
    mask = get_sam_mask_from_box(frame_rgb, best_box).astype(np.uint8)
    return mask


def run_change_color_gradual_pipeline(
    input_video: str,
    output_video: str,
    target_prompt: str,
    target_color: str,
    dino_box_threshold: float = 0.30,
    dino_text_threshold: float = 0.25,
    max_frames: int | None = None,
) -> dict:
    """Run color change with gradual hue shift across frames.
    
    Color shifts gradually from frame 0 to last frame.
    """
    from_color = 'black'
    hue_shift_final = estimate_hue_shift(from_color, target_color)
    
    frames, fps, width, height = load_video(input_video)
    if max_frames is not None:
        frames = frames[:max_frames]
    
    out_frames: list[np.ndarray] = []
    num_frames = len(frames)
    
    for frame_idx, frame in enumerate(tqdm(frames, desc='change_color_gradual')):
        progress = frame_idx / max(num_frames - 1, 1)
        current_hue_shift = int(hue_shift_final * progress)
        
        target_mask = estimate_target_mask_gdino_sam(
            frame_bgr=frame,
            target_prompt=target_prompt,
            dino_box_threshold=dino_box_threshold,
            dino_text_threshold=dino_text_threshold,
        )
        
        edited = apply_color_change(
            frame_bgr=frame,
            target_mask_u8=target_mask,
            hue_shift=current_hue_shift,
            saturation_scale=1.0 + 0.2 * progress,
            value_scale=1.0,
        )
        out_frames.append(edited)
    
    write_video(output_video, out_frames, fps, width, height)
    
    return {
        'input_video': input_video,
        'output_video': output_video,
        'target_prompt': target_prompt,
        'target_color': target_color,
        'hue_shift_max': hue_shift_final,
        'fps': fps,
        'width': width,
        'height': height,
        'num_frames': num_frames,
        'gradual': True,
    }


print('Ready: change_color pipeline with gradual mode')
```
